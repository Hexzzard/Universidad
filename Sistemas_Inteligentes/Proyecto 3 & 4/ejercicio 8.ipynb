```python
#codigo de la pregunta 6
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris

iris = load_iris() #cargamos el dataset

#convertimos el dataset a un Dataframe de pandas
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']])
df.columns = iris['feature_names'] + ['target']

kmeans = KMeans(n_clusters=3).fit(df[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']])
clase = kmeans.labels_  #obtenemos las clases de cada uno de los puntos

#hacemos lo mismo que el anterior, guardamos dentro de un arreglo los colores de cada uno de los puntos
Select = []
color = ['c','yellow','purple']
for i in clase:
    Select.append(color[i])
fig = plt.figure()
df.plot("sepal length (cm)", "sepal width (cm)", color=Select, kind="scatter")
plt.title("Clases predichas obtenidas de Kmeans (pregunta 6)")
plt.show()
```

```python
#Ahora lo que debemos hacer es obtener los centroides de los cluster obtenidos desde Kmeans
#esto lo obtenemos mediante el atributo "cluster_centers_"
centroides = kmeans.cluster_centers_

#graficamos nuevamente
fig = plt.figure()
df.plot("sepal length (cm)", "sepal width (cm)", color=Select, kind="scatter")
plt.scatter(centroides[:,0], centroides[:,1], marker='*', c='black',s=100)
plt.title("Clases predichas obtenidas de Kmeans con centroides graficados")

#mostramos por pantalla los centroides obtenidos
for i in range(len(centroides)):
    print("Centroide",i,"=",centroides[i])
plt.show()
```
