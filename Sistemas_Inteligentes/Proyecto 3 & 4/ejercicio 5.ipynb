```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans

from mpl_toolkits.mplot3d import axes3d
from matplotlib import style

import warnings
warnings.filterwarnings('ignore') #ignoramos las adevertencias
#Nota: se produce una debido al procesamiento de la cantidad de hilos, lo cual nos suguiere cambiar la configuracion de hilos

#obtenemos los datos desde el archivo "puntos_3d.npy"
d3 = np.load('puntos_3d.npy')

#generamos un dataframe con los datos obtenidos
df = pd.DataFrame(data=d3)
df.columns = ['x', 'y', 'z']
```

```python
#graficamos los datos obtenidos
#nota: los datos se encuentran en 3d

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection='3d')

#graficamos los datos
ax.scatter(df.iloc[: , 0].values, df.iloc[: , 1].values, df.iloc[: , 2].values, c='red',s=60)
plt.title('Sin centroide')
plt.show()
```

```python
#utilizaremos Kmeans para obtener el centroide, como queremos el centroide de todos los datos
#el valor de K sera igual a 1
kmeans = KMeans(n_clusters=1).fit(df)
#con el atributo "cluster_centers_", obtendremos el centroide de la figura
centroide = kmeans.cluster_centers_
print("centroide= ",centroide[0])
```

```python
#Ahora que hemos determinado el centroide graficamos nuevamente
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection='3d')
ax.scatter(df.iloc[: , 0].values, df.iloc[: , 1].values, df.iloc[: , 2].values, c='red',s=60)

#graficamos el centroide
ax.scatter(centroide[0][0], centroide[0][1], centroide[0][2], c='blue',s=300)
plt.title('Con centroide')
plt.show()
```
