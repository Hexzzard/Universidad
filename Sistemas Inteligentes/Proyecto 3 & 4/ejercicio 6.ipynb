```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris

iris = load_iris() #cargamos el dataset

#convertimos el dataset a un Dataframe de pandas
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']])
df.columns = iris['feature_names'] + ['target']

df
```

```python
#generaremos un grafico con los datos del dataset, en donde podamos visualizar la clase original
#del dataset, para compararlo con lo obtenido con Kmeans
color = ['c','yellow','purple']

#generaremos un arreglo el cual contendra los colores que tendra el punto en funcion de su clase
Select = []
for i in df.iloc[: , 4].values:
    Select.append(color[int(i)])
fig = plt.figure()
df.plot("sepal length (cm)", "sepal width (cm)", color=Select, kind="scatter")
plt.title("Datos reales obtenidos desde el dataset Iris")
plt.show()
```

```python
#ahora utilizaremos Kmeans para comparar la agrupacion de clases generada con la original

#generamos nuestro Kmeans
kmeans = KMeans(n_clusters=3).fit(df[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']])
clase = kmeans.labels_  #obtenemos las clases de cada uno de los puntos

#hacemos lo mismo que el anterior, guardamos dentro de un arreglo los colores de cada uno de los puntos
Select = []
color = ['c','yellow','purple']
for i in clase:
    Select.append(color[i])
labels
```
